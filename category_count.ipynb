```python
import os

path ="/content/drive/My Drive/fina_fina/labels"

# Change the directory
os.chdir(path)

def read_files(file_path):
   with open(file_path, 'r') as file:
      #print(file.read())
      for line in file:
         print(line[0])
         if line[0] == '0' :
           s_unripe0.append(line[0])
         elif line[0] == '1' :
           s_pripe1.append(line[0])
         elif line[0] == '2' :
           s_ripe2.append(line[0])
         elif line[0] == '3' :
           s_rotten3.append(line[0])
         elif line[0] == '4' :
           a_unripe4.append(line[0])
         elif line[0] == '5' :
           a_pripe5.append(line[0])
         elif line[0] == '6' :
           a_ripe6.append(line[0])
         elif line[0] == '7' :
           a_rotten7.append(line[0]) 
         else :
           continue

# Iterate all the files in the directory
for file in os.listdir():
   if file.endswith('.txt'):
      # Create the filepath of particular file
      file_path =f"{path}/{file}"
      read_files(file_path)
      print(file_path)
```
